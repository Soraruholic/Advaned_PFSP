```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib as mpl
%matplotlib inline
plt.rcParams['font.sans-serif'] = ['SimHei']       #解决中文乱码
plt.rcParams['axes.unicode_minus'] = False      #解决负号不显示
```

```python
large = 22; med = 16; small = 12
params = {'axes.titlesize': 40, #子图上的标题字体大小
          'legend.fontsize': 30, #图例的字体大小
          'figure.figsize': (20, 20), #图像的画布大小
          'axes.labelsize': 30, #标签的字体大小
          'xtick.labelsize': 30, #x轴上的标尺的字体大小
          'ytick.labelsize': 30, #y轴上的标尺的字体大小
          'figure.titlesize': large} #整个画布的标题字体大小
plt.rcParams.update(params) #设定各种各样的默认属性
 # 设定整体风格
plt.style.use('seaborn-whitegrid')
# 设定整体背景风格
sns.set_style("white")
```

```python
datas = pd.read_csv ('Climb_rec_tune_MaxLoops.csv')
datas2 = pd.read_csv ('PMCL_rec_tune_MaxLoops.csv')
```

```python
data = pd.DataFrame({"iterations":datas['ID'],"error":datas['RPD']})
data2 = pd.DataFrame({"iterations":datas2['ID'],"error":datas2['RPD']})   # 必须把所有数据放到DataFrame里面
sns.regplot("iterations","error",data=data,
                      marker = '*',
                      logx=True,
                      line_kws={'label': 'Climb'},
                      scatter_kws = dict(s = 6, linewidths = 0.7, edgecolors = 'black')) # 散点大小、线宽、点边缘色

sns.regplot("iterations","error",data=data2,
                      marker = '*',
                      logx=True,
                      line_kws={'label': 'Climb+Palmer'},
                      scatter_kws = dict(s = 6, linewidths = 0.7, edgecolors = 'black'))                      
sns.despine(top=False, right=False, left=False, bottom=False) # 设置边框是否显示
plt.xticks(fontsize=30,fontproperties = 'Times New Roman')   # 设置x轴刻度大小和字体类型
plt.yticks(fontsize=30,fontproperties = 'Times New Roman')
plt.tick_params(bottom=True,left=True,right=False,top=False)
plt.legend(loc='best')
plt.rcParams['xtick.direction']='out'   # x轴的刻度线向内显示
```
